What are some limitations or challenges of Association rules mining?

Sparsity: In large datasets with many items, most itemsets will not appear together frequently, making it challenging to extract meaningful rules.
Threshold selection: Setting the appropriate support, confidence, and lift thresholds can be tricky. Too high or too low can either miss important rules or result in too many uninteresting ones.
Scalability: The computational cost increases significantly with the size of the dataset, especially when the number of items is large.
