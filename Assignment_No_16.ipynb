# Grid search for Algorithm Tuning

```python
n_neighbors = np.array(range(1,40))
param_grid = dict(n_neighbors=n_neighbors)
param_grid
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator=model, param_grid=param_grid,cv=10)
grid.fit(X_train, Y_train)
print(grid.best_params_)
```

##### After applying GridSearch, we got the best K (n_neighbors) value as 1, so we will be using the k= 1 for KNN Classifier algorithm

### Visualizing CV results

```python
import matplotlib.pyplot as plt
%matplotlib inline
# choose k between 1 to 41
k_range = range(1, 41)
k_scores = []
# use iteration to caclulator different k in models, then return the average accuracy based on the cross validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, Y_train, cv=10)
    k_scores.append(scores.mean())
# plot to see clearly
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

##### We could see that the model accuracy is very good for k values smaller than 5 and as the value increases the accuracy goes on decreasing

# Using KNN Classifier for prediction

```python
model = KNeighborsClassifier(n_neighbors =1).fit(X_train,Y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test,y_pred)
print(accuracy)
```

##### We can see that the accuracy score which we have got for our model is 0.76 which is 76%. It is decent accuracy score. But the accuracy score can be misleading for imbalanced data. So we will use confusion matrix and classification report metrics further

```python
confusion_matrix = confusion_matrix(Y_test,y_pred)
print (confusion_matrix)
```

```python
print(classification_report(Y_test,y_pred))
```

##### The precison and f1 score for type 5 is  low. Since the data is imbalanced, we can see the precision values are affected. We will use oversamping technique as the data is very less and undersampling will cause data loss

```python
import matplotlib.pyplot as plt
%matplotlib inline
# choose k between 1 to 41
k_range = range(1, 41)
k_scores = []
# use iteration to caclulator different k in models, then return the average accuracy based on the cross validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, Y_train, cv=10)
    k_scores.append(scores.mean())
# plot to see clearly
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

### 1. What are the key hyperparameters in KNN?  
The key hyperparameters in the **K-Nearest Neighbors (KNN)** algorithm include:  

- **k (Number of Neighbors):** Determines how many nearest neighbors are considered for classification or regression. A small k may lead to overfitting, while a large k can smooth the decision boundary but may cause underfitting.  
- **Distance Metric:** Defines how the distance between data points is calculated (e.g., Euclidean, Manhattan, Minkowski, etc.).  
- **Weighting Function:** Determines how neighbors contribute to the prediction. Common options include:
  - **Uniform weighting:** All neighbors contribute equally.  
  - **Distance weighting:** Closer neighbors have higher influence.  
- **Algorithm for Finding Neighbors:** Specifies how neighbors are searched efficiently. Options include:
  - **Brute Force:** Computes distances directly for all points.  
  - **KD-Tree:** Efficient for low-dimensional data.  
  - **Ball-Tree:** Suitable for higher-dimensional data.  
- **Leaf Size:** Used in KD-Tree and Ball-Tree to control the number of points in a leaf node, affecting speed and memory usage.

### 2. What distance metrics can be used in KNN?  
Several distance metrics can be used in **KNN** to measure the similarity between data points:  

- **Euclidean Distance:**  
  \[
  d(p, q) = \sqrt{\sum (p_i - q_i)^2}
  \]  
  The most commonly used metric, suitable for continuous numerical data.  

- **Manhattan Distance (L1 norm or Taxicab distance):**  
  \[
  d(p, q) = \sum |p_i - q_i|
  \]  
  Useful when movement is constrained to grid-like paths.  

- **Minkowski Distance:** A generalized form of Euclidean and Manhattan distances:  
  \[
  d(p, q) = \left(\sum |p_i - q_i|^p \right)^{\frac{1}{p}}
  \]  
  - When \( p = 1 \), it is Manhattan distance.  
  - When \( p = 2 \), it is Euclidean distance.  

- **Chebyshev Distance:**  
  \[
  d(p, q) = \max |p_i - q_i|
  \]  
  Used when movement is only allowed in horizontal, vertical, or diagonal directions.  

- **Hamming Distance:** Measures the number of positions at which two binary vectors differ, used for categorical data.  

- **Cosine Similarity (1 - Cosine Distance):**  
  \[
  d(p, q) = 1 - \frac{p \cdot q}{\|p\| \|q\|}
  \]  
  Measures the angle between two vectors, often used in text mining and recommendation systems.  

- **Mahalanobis Distance:**  
  \[
  d(p, q) = \sqrt{(p - q)^T S^{-1} (p - q)}
  \]  
  Accounts for correlations between variables and is useful when features have different scales.
